```python
import pandas as pd
import numpy as np
from Code.preprocessing.netflix_data import load_from_txt
import matplotlib.pyplot as plt
```

```python
df = load_from_txt('Data')
#lol says decompressing files... 
# it's because it's using the default progress reporter for the file decompression routine... 
# not worth fixing...
```

```python
user = df.groupby('user_id').agg(['count'])
```

```python
user.head()
```

```python
user_counts = user.loc[:,('rating','count')]
user_counts.head()
```

```python
plt.hist(user_counts,bins=100)
plt.ylabel('Number of Users')
plt.xlabel('Number of Reviews')
plt.title('Histogram of Number of Ratings per User (All)')
```

```python
# lets 'zoom in', and only look at histogram for <1000 reviews/user
plt.hist(user_counts[user_counts<1000],bins=100)
plt.ylabel('Number of Users')
plt.xlabel('Number of Reviews')
plt.title('Histogram of Number of Ratings per User (<1000)')
```

```python
# lets 'zoom in', and only look at histogram for <100 reviews/user
plt.hist(user_counts[user_counts<100],bins=50)
plt.ylabel('Number of Users')
plt.xlabel('Number of Reviews')
plt.title('Histogram of Number of Ratings per User (<100)')
```
